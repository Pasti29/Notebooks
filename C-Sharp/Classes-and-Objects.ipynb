## Classes and Objects
---------

#### Theory

<span style="color:purple;">Class</span> is a template or blueprint from which <span style="color:purple;">objects</span> are created. A class provides initial values for *states* (member variables or attributes) and implementations of *behaviors* (member functions or methods). An **instance** is a specific object created from a particular class. Classes supports **inheritance**.

The image above shows a `Car` class that is a template for many `Car` instances. There are three instances: `polo`, `mini` and `beetle`.

#### Example syntax

```
class Car
{
    private string _model;
    private string _color;
    private int _passengers;
    private int _speed;

    public Car(string model, string color, int passengers, int speed)
    {
        _model = model;
        _passengers = passengers;
        _color = color;
        _speed = speed;
    }

    public void Accelerate()
    {
        _speed += 2;
    }
}
```

```
// One way of creating an object
var bmw = new Car("BMW", "red", 4, 5);

// Second way of creating an object
Car ferrari = new("Ferrari", "white", 2, 10);
```
